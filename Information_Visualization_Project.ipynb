##Import & Load all the Audio Files

```python
#Raw File
raw_file = '/content/drive/MyDrive/IVemc689Audio/RawWav-Edit.wav'
raw, srraw  = librosa.load(raw_file)

#Air File
air_file = '/content/drive/MyDrive/IVemc689Audio/AirWav-Edit.wav'
air, srair  = librosa.load(air_file)

#Narrow File
narrow_file = '/content/drive/MyDrive/IVemc689Audio/NarrowWav-Edit.wav'
nar, srnar  = librosa.load(narrow_file)

#Room 1 File
room1_file = '/content/drive/MyDrive/IVemc689Audio/Room1Wav-Edit.wav'
r1, sr1  = librosa.load(room1_file)

#Room 2 File
room2_file = '/content/drive/MyDrive/IVemc689Audio/Room2WavEdit.wav'
r2, sr2  = librosa.load(room2_file)

#Room 3 File
room3_file = '/content/drive/MyDrive/IVemc689Audio/Room3Wav-Edit.wav'
r3, sr3  = librosa.load(room3_file)
```

##Sound Wave Graph and Mel Spectrograms

###Raw Audio

```python
#Plot the sound waves using Amplitude and Time
plt.figure(figsize=(14, 5))
plt.title('Raw Audio Sound Wave')
plt.ylabel('Amplitude (dB)')
plt.xlabel('Time(s)')
librosa.display.waveplot(raw, sr=srraw)
```

```python
#Convert the data into spectrogram
raw_spec = librosa.feature.melspectrogram(y=raw, sr=srraw)
rawS_DB = librosa.power_to_db(raw_spec, ref=np.max)
librosa.display.specshow(rawS_DB, sr=srraw, x_axis='time', y_axis='mel');
plt.colorbar(format='%+2.0f dB');
```
